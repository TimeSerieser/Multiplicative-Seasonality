```python
import numpy as np
import pandas as pd

df = pd.read_csv('MA.csv')
df = df.dropna()
df.head()
```

```python
df.tail()
```

```python
alpha = 0.22
gamma = 0.64
s = 4
df['Level'] = 0.00
df['Level'][s-1] = np.mean([df['Y_true'][0:s]])
df['Level'][s-1]
```

```python
df['Seasonality'] = 0.00

for i in range(0,s):
    df['Seasonality'][i] = df['Y_true'][i] / df['Level'][s-1]

df.head() 
```

```python
for i in range(s,len(df.Y_true)):
    df['Level'][i] = alpha * ( df['Y_true'][i] / df['Seasonality'][i-s] ) + (1-alpha) * df['Level'][i-1]
    df['Seasonality'][i] = gamma * ( df['Y_true'][i] / df['Level'][i] ) + (1-gamma) * df['Seasonality'][i-s]

df.head(8)
```

```python
df['Y_pred'] = 0.00

for i in range(s,len(df.Y_true)):
    df['Y_pred'][i] =  df['Level'][i-1] * df['Seasonality'][i-s]

df.head(8)  
```

```python
df.tail(8)
```

```python
df['Absolute Error'] = 0.00

for i in range(s,len(df.Y_true)):
    df['Absolute Error'][i] = abs( df['Y_pred'][i] - df['Y_true'][i] )

df.head(8)                                
```

```python
df.tail(8)
```

```python
df['Squared Error'] = 0.00

for i in range(s,len(df.Y_true)):
    df['Squared Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )**2

df.head(8)                                
```

```python
df.tail(8) 
```

```python
df['Precent Error1'] = 0.00

for i in range(s,len(df.Y_true)):
    df['Precent Error1'][i] = abs ((  df['Y_true'][i] - df['Y_pred'][i] ) / df['Y_true'][i])

df.head(8)                                
```

```python
df.tail(8)
```

```python
df['Precent Error2'] = 0.00

for i in range(s,len(df.Y_true)):
    df['Precent Error2'][i] = ( (  df['Y_pred'][i] - df['Y_true'][i] ) / df['Y_true'][i-1] )**2

df.head(8)                                
```

```python
df.tail(8) 
```

```python
df['Precent Error3'] = 0.00

for i in range(s,len(df.Y_true)):
    df['Precent Error3'][i] = ( (  df['Y_true'][i] - df['Y_true'][i-1] ) / df['Y_true'][i-1] )**2

df.head(8)                                
```

```python
df.tail(8) 
```

```python
df['Error'] = 0.00

for i in range(s,len(df.Y_true)):
    df['Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )

df.head(8)                                
```

```python
df.tail(8) 
```

```python
df['Squared Error Diff'] = 0.00

for i in range(s+1,len(df.Y_true)):
    df['Squared Error Diff'][i] = ( df['Error'][i] - df['Error'][i-1] )**2

df.head(9)                                
```

```python
df.tail(9)
```

```python
MSE = np.mean(df['Squared Error'][s:-1])
MSE
```

```python
RMSE = np.sqrt(np.mean(df['Squared Error'][s:-1]))
RMSE
```

```python
MAE = np.mean(df['Absolute Error'][s:-1])
MAE
```

```python
MAPE = np.mean(df['Precent Error1'][s:-1])
MAPE
```
